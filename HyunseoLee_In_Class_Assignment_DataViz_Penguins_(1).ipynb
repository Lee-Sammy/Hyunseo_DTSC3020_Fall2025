## 0) Setup

```python
# Run it first.
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

sns.set_theme(style='whitegrid')
penguins = sns.load_dataset('penguins')
penguins = penguins.dropna()
penguins.head()  # preview
```

## 1–3) Scatter plot with color by species
Create a scatter plot with:
- **x =** `bill_length_mm`
- **y =** `bill_depth_mm`
- Color points by **species**

Then add a short comment (2–3 sentences) describing the pattern(s) you see.

```python
# TODO: Create the scatter plot here
# 1–3) Scatter plot with color by species
plt.figure(figsize=(8,6))
sns.scatterplot(
    data=penguins,
    x='bill_length_mm',
    y='bill_depth_mm',
    hue='species',
    palette='Set2'
)
plt.title("Penguin Bill Length vs. Bill Depth by Species")
plt.xlabel("Bill Length (mm)")
plt.ylabel("Bill Depth (mm)")
plt.show()

```

**Your interpretation (2–3 sentences):**

_Write your observations here._

From the scatter plot, we can see that each penguin species has a distinct range of bill sizes.
Gentoo penguins generally have the longest bills but shallower depths, while Adelie penguins have shorter and deeper bills.
Chinstrap penguins fall between them, showing moderate length and depth, which helps visually separate the three species.

## 4) Histogram with two different bin sizes
Plot a histogram of `bill_length_mm` **twice**:
- Once with **10 bins**
- Once with **30 bins**

Then write 1–2 sentences comparing what changes when you change the number of bins.

```python
# TODO: Histogram with 10 bins
plt.figure(figsize=(8,5))
sns.histplot(data=penguins, x='bill_length_mm', bins=10, color='skyblue', edgecolor='black')
plt.title("Histogram of Bill Length (10 bins)")
plt.xlabel("Bill Length (mm)")
plt.ylabel("Count")
plt.show()

```

```python
# TODO: Histogram with 30 bins
plt.figure(figsize=(8,5))
sns.histplot(data=penguins, x='bill_length_mm', bins=30, color='salmon', edgecolor='black')
plt.title("Histogram of Bill Length (30 bins)")
plt.xlabel("Bill Length (mm)")
plt.ylabel("Count")
plt.show()
```

**Your comparison (1–2 sentences):**

_Write your observations here._

With 10 bins, the distribution looks smoother and easier to interpret, but small details are hidden.
With 30 bins, more fine-grained variation appears, revealing smaller peaks but making the plot look noisier.

## 5) Boxplot of body mass by species
Create a **boxplot** showing `body_mass_g` grouped by `species`.
Then write 1–2 sentences summarizing which species are heavier/lighter on average.

```python
# TODO: Boxplot of body mass by species
plt.figure(figsize=(8,6))
sns.boxplot(
    data=penguins,
    x='species',
    y='body_mass_g',
    palette='Set3'
)
plt.title("Penguin Body Mass by Species")
plt.xlabel("Species")
plt.ylabel("Body Mass (g)")
plt.show()

```

**Your summary (1–2 sentences):**

_Write your observations here._

Gentoo penguins are the heaviest on average, with body masses generally around 5,000 g or more.
Adelie penguins are the lightest, while Chinstrap penguins fall in between, showing moderate weight levels.
